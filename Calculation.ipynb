```python
import sys
import os
import math
from numpy import *
from sympy import *
```

```python
lines = []
vectors = []
angles = []
cNumbers = []

try:
    open('=.pipe')
except FileNotFoundError:
    sys.exit("File not found.")

with open ('=.pipe') as pipe:
    for line in pipe:
        lines.append(line.rstrip('\n'))
pipe.close()

    
atom = input("Please enter the atom number(s) need to be calculated, seperate with space: ")
atomStr = atom.split()
iterations = len(atomStr)
i = 0

while i < iterations:
    atomNum = int(atomStr[i])
    
    vectorData = lines[7]
    angleData = lines[8]
    vectors = vectorData.split()
    angles = angleData.split()

    a = float(vectors[1])
    b = float(vectors[2])
    c = float(vectors[3])

    alpha = math.radians(float(angles[1]))
    beta = math.radians(float(angles[2]))
    gamma = math.radians(float(angles[3]))


    Calcny1 = round(math.sqrt(2)*(cos(alpha)*cot(beta)-cos(gamma)*csc(beta)) / math.sqrt(-2*(cos(alpha))**2-cos(2*beta)+4*cos(alpha)*cos(beta)*cos(gamma)-cos(2*gamma)),10)
    Calcny2 = round(math.sqrt(2)*sin(beta) / math.sqrt(-2*(cos(alpha))**2-cos(2*beta)+4*cos(alpha)*cos(beta)*cos(gamma)-cos(2*gamma)),10)
    Calcny3 = round(math.sqrt(2)*(cos(gamma)*cot(beta)-cos(alpha)*csc(beta)) / math.sqrt(-2*(cos(alpha))**2-cos(2*beta)+4*cos(alpha)*cos(beta)*cos(gamma)-cos(2*gamma)),10)
    Calcnz1 = round(-cot(beta), 10)
    Calcnz3 = round(csc(beta), 10)

    nx1 = 1
    nx2 = 0
    nx3 = 0
    ny1 = float("{:.10f}".format(float((f"{Calcny1:.10f}"))))
    ny2 = float("{:.10f}".format(float((f"{Calcny2:.10f}"))))
    ny3 = float("{:.10f}".format(float((f"{Calcny3:.10f}"))))
    nz1 = float("{:.10f}".format(float((f"{Calcnz1:.10f}"))))
    nz2 = 0
    nz3 = float("{:.10f}".format(float((f"{Calcnz3:.10f}"))))


    if atomNum <= len(lines) - 14:
        atomNumLine = 9 + atomNum
        cNumbersData = lines[atomNumLine]
        cNumbers = cNumbersData.split()
        c1 = float(cNumbers[3])
        c2 = float(cNumbers[4])
        c3 = float(cNumbers[5])
    
        Pcx1 = f"{(c1 + 0.001*nx1/a):.10f}"
        Pcx2 = f"{(c2 + 0.001*nx2/b):.10f}"
        Pcx3 = f"{(c3 + 0.001*nx3/c):.10f}"
        Ncx1 = f"{(c1 - 0.001*nx1/a):.10f}"
        Ncx2 = f"{(c2 - 0.001*nx2/b):.10f}"
        Ncx3 = f"{(c3 - 0.001*nx3/c):.10f}"
    
        Pcy1 = f"{(c1 + 0.001*ny1/a):.10f}"
        Pcy2 = f"{(c2 + 0.001*ny2/b):.10f}"
        Pcy3 = f"{(c3 + 0.001*ny3/c):.10f}"
        Ncy1 = f"{(c1 - 0.001*ny1/a):.10f}"
        Ncy2 = f"{(c2 - 0.001*ny2/b):.10f}"
        Ncy3 = f"{(c3 - 0.001*ny3/c):.10f}"
    
        Pcz1 = f"{(c1 + 0.001*nz1/a):.10f}"
        Pcz2 = f"{(c2 + 0.001*nz2/b):.10f}"
        Pcz3 = f"{(c3 + 0.001*nz3/c):.10f}"
        Ncz1 = f"{(c1 - 0.001*nz1/a):.10f}"
        Ncz2 = f"{(c2 - 0.001*nz2/b):.10f}"
        Ncz3 = f"{(c3 - 0.001*nz3/c):.10f}"
    
    else:
        print('Atom' + atomNum + 'does not exist')
    
    
    parent_dir = "./"
    directory1 = "dis_" + str(atomNum) + "_0_0_-0.001" 
    directory2 = "dis_" + str(atomNum) + "_0_0_0.001"
    directory3 = "dis_" + str(atomNum) + "_0_-0.001_0"
    directory4 = "dis_" + str(atomNum) + "_0_0.001_0"
    directory5 = "dis_" + str(atomNum) + "_-0.001_0_0"
    directory6 = "dis_" + str(atomNum) + "_0.001_0_0"
    path1 = os.path.join(parent_dir, directory1)
    path2 = os.path.join(parent_dir, directory2)
    path3 = os.path.join(parent_dir, directory3)
    path4 = os.path.join(parent_dir, directory4)
    path5 = os.path.join(parent_dir, directory5)
    path6 = os.path.join(parent_dir, directory6)
    try:
        os.mkdir(path1)
        os.mkdir(path2)
        os.mkdir(path3)
        os.mkdir(path4)
        os.mkdir(path5)
        os.mkdir(path6)
        print("successfully created directories for atom: " + str(atomNum))
    except OSError as error:
        print(error)
    
    os.chdir(path1)
    out1 = open("=.pipe", "w")
    linesNew = lines.copy()
    change = linesNew[atomNumLine].split()
    change[3] = str(Ncz1)
    change[4] = str(Ncz2)
    change[5] = str(Ncz3)
    str1 = " ".join(change)
    linesNew[atomNumLine] = str1
    for line in linesNew:
        out1.write(str(line))
        out1.write("\n")
    out1.close()
    
    os.chdir('.'+path2)
    out2 = open("=.pipe", "w")
    change[3] = str(Pcz1)
    change[4] = str(Pcz2)
    change[5] = str(Pcz3)
    str2 = " ".join(change)
    linesNew[atomNumLine] = str2
    for line in linesNew:
        out2.write(str(line))
        out2.write("\n")
    out2.close()
    
    os.chdir('.'+path3)
    out3 = open("=.pipe", "w")
    change[3] = str(Ncy1)
    change[4] = str(Ncy2)
    change[5] = str(Ncy3)
    str3 = " ".join(change)
    linesNew[atomNumLine] = str3
    for line in linesNew:
        out3.write(str(line))
        out3.write("\n")
    out3.close()
    
    os.chdir('.'+path4)
    out4 = open("=.pipe", "w")
    change[3] = str(Pcy1)
    change[4] = str(Pcy2)
    change[5] = str(Pcy3)
    str4 = " ".join(change)
    linesNew[atomNumLine] = str4
    for line in linesNew:
        out4.write(str(line))
        out4.write("\n")
    out4.close()
    
    os.chdir('.'+path5)
    out5 = open("=.pipe", "w")
    change[3] = str(Ncx1)
    change[4] = str(Ncx2)
    change[5] = str(Ncx3)
    str5 = " ".join(change)
    linesNew[atomNumLine] = str5
    for line in linesNew:
        out5.write(str(line))
        out5.write("\n")
    out5.close()
    
    os.chdir('.'+path6)
    out6 = open("=.pipe", "w")
    change[3] = str(Pcx1)
    change[4] = str(Pcx2)
    change[5] = str(Pcx3)
    str6 = " ".join(change)
    linesNew[atomNumLine] = str6
    for line in linesNew:
        out6.write(str(line))
        out6.write("\n")
    out6.close()
    
    i += 1
    
 
```
